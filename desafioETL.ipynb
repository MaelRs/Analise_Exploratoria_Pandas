```python
# Criação de função para verificar a atualização do usuário
def update_user(user, df):
    mensagem_col = df['Mensagem']
    if mensagem_col.str.contains(user).any():
        print(f"Usuário {user} Atualizado com sucesso!")
    else:
        print(f"Não foi possível atualizar o usuário {user}.")

# Percorrer a lista de usuários
for user in users_ids:
    update_user(user, df_client)
```
