```python
import pandas as pd
# Projeto realizado integralmente com extração de dados de arquivo CSV, devido a indisponibilidade das aplicações sugeridas pelo Professor durante a aula como Swagger, e API da SDW-2023.
#Repositório da API: https://github.com/digitalinovationone/santander-dev-week-2023-api
sdw2023_api_url = 'https://sdw-2023-prd.up.railway.app/api/users'


# Arquivo CSV para geração de lista de clientes.
df_client=pd.read_csv('/content/drive/MyDrive/SDW2023.csv')
users_ids = df_client['Name'].tolist()
print(users_ids)
#df.head()
```

```python
#Arquivo CSV para gerar lista de mensagens
df_0=pd.read_csv('/content/drive/MyDrive/Mensagens.csv')
msg_ids = df_0['Mensagem'].tolist()
news=[]
for mensage in msg_ids:
  news.append(mensage)
#print(news)
```

```python
# Arquivo CSV para geração da lista de saudações
df_saudacao=pd.read_csv('/content/drive/MyDrive/Saudacoes.csv')
sds_ids = df_saudacao['saudacao'].tolist()
sds=[]
for saudacao in sds_ids:
  sds.append(saudacao)
```

```python
#Função para geração das mensagens aleatórias combinadas com cada usuário
from random import randint
from time import sleep
msg_list=[]
def generate_ai_news(user):
  msg=randint(0,len(news)-1)
  saud=randint(0,len(sds)-1)
  mensagem=f'{sds[saud]} {user}, {news[msg]}'
  msg_list.append(mensagem)
  print(mensagem)
  print()
  sleep(.5)

for user in users_ids:
  mensagem = generate_ai_news(user)

```

```python
# Cria a coluna Mensagem no data frame e Carrega as mensagens
df_client['Mensagem']=msg_list
```

```python
# Criação de função para verificar a atualização do usuário
def update_user(user, df):
    mensagem_col = df['Mensagem']
    if mensagem_col.str.contains(user).any():
        print(f"Usuário {user} Atualizado com sucesso!")
    else:
        print(f"Não foi possível atualizar o usuário {user}.")

# Percorrer a lista de usuários
for user in users_ids:
    update_user(user, df_client)
```
